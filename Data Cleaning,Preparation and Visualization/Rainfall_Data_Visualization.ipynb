## Data Visualization on Rainfall Dataset.

## Importing Required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Reading Csv file

## This dataset contains the Rainfall data.
### Dataset is taken from: https://data.gov.in/resource/rainfall-south-peninsula-10-subdivisions-and-its-departure-normal-monsoon-session-june-0

```python
df = pd.read_csv("/content/south_pen-India_rainfall_act_dep_1901_2016.csv")

# Printing First 7 rows of dataset
print("Dataset is:\n")
print(df.head(n = 7))
```

## Seeing the Columns of Dataset

```python
print("Columns of Dataset are:\n", df.columns)

print("\nShape of Dataset is:", df.shape)
```

## Checking the Data types of Columns

```python
# Total Count of Datatypes
print("Datatype of Dataset is:")
df.info()

# According to Columns
data_types = df.dtypes

print("\nDatatypes of Columns is:\n")
print(data_types)
```

## Seeing Descriptive Statics of Dataset

```python
print("Descriptive Statics of Dataset are:\n")
df.describe(include = 'all')
```

## Checking for Null Values in Dataset

```python
print("Count of Null Values in Dataset is:\n")
df.isnull().sum()
```

## Data Visualization

## Visualizing Scatter plot Actual Rainfall for each month with Year.

```python
for i, col in enumerate(df.columns):

  # Plotting Scatter Plot
  if i != 1 and i < len(df.columns) - 5 and col != 'YEAR':

    sns.scatterplot(x = 'YEAR', y = col, data = df)
    # Adding Labels
    plt.xlabel("Year")
    plt.ylabel(f"{col}")
    plt.title(f'Scatter Plot of {col} against YEAR')
    plt.show()
```

## Visualizing Scatter plot Departure Percentage for each month with Year.

```python
for i, col in enumerate(df.columns):

  # Plotting Scatter Plot
  if i != 1 and i > len(df.columns) - 5 and col != 'YEAR':

    sns.scatterplot(x = 'YEAR', y = col, data = df)
    # Adding Labels
    plt.xlabel("Year")
    plt.ylabel(f"{col}")
    plt.title(f'Scatter Plot of {col} against YEAR')
    plt.show()
```

## Bar Plot of average Actual Rainfall for each month.

```python
# Calculate the mean for each month
average_rainfall_by_month = df.drop(columns =['YEAR', 'Departure Percentage: JUN', 'Departure Percentage: JUL', 'Departure Percentage: AUG', 'Departure Percentage: SEP', 'Departure Percentage: JUN-SEPT'], axis=1).mean()

# Plotting the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=average_rainfall_by_month.index, y=average_rainfall_by_month.values)
plt.xlabel("Month")
plt.ylabel("Average Rainfall")
plt.title("Average Rainfall for Each Month")
plt.xticks(rotation=45)
plt.show()
```

## Bar Plot of average Departure Percentage for each month.

```python
# Calculate the mean for each month
df_dropYear = df.drop('YEAR', axis = 1)
col = ['Departure Percentage: JUN', 'Departure Percentage: JUL', 'Departure Percentage: AUG', 'Departure Percentage: SEP', 'Departure Percentage: JUN-SEPT']
average_departure_by_month = df_dropYear[col].mean()

# Plotting the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=average_departure_by_month.index, y=average_departure_by_month.values)
plt.xlabel("Month")
plt.ylabel("Average Departure Percentage")
plt.title("Average Departure Percentage for Each Month")
plt.xticks(rotation=45)
plt.show()
```

## Line Plot for Actual Rainfall Over the Years.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='YEAR', y='Actual Rainfall: JUN', data=df, label='JUN')
sns.lineplot(x='YEAR', y='Actual Rainfall: JUL', data=df, label='JUL')
sns.lineplot(x='YEAR', y='Actual Rainfall: AUG', data=df, label='AUG')
sns.lineplot(x='YEAR', y='Actual Rainfall: SEPT', data=df, label='SEPT')
plt.title('Actual Rainfall Over the Years')
plt.xlabel('Year')
plt.ylabel('Actual Rainfall')
plt.legend()
plt.show()
```

## Line Plot for Departure Percentage Over the Years.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='YEAR', y='Departure Percentage: JUN', data=df, label='JUN')
sns.lineplot(x='YEAR', y='Departure Percentage: JUL', data=df, label='JUL')
sns.lineplot(x='YEAR', y='Departure Percentage: AUG', data=df, label='AUG')
sns.lineplot(x='YEAR', y='Departure Percentage: SEP', data=df, label='SEPT')
plt.title('Departure Percentage Over the Years')
plt.xlabel('Year')
plt.ylabel('Departure Percentage')
plt.legend()
plt.show()
```

## Box Plot of Actual Rainfall for Months.

```python
rainfall_columns = ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT', 'Actual Rainfall: JUN-SEPT']

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[rainfall_columns])
plt.title('Box Plot of Actual Rainfall for Months')
plt.ylabel('Actual Rainfall')
plt.xticks(rotation = 45)
plt.xlabel("Months")
plt.show()
```

## Box Plot for Departure Percentages for Months.

```python
departure_columns = ['Departure Percentage: JUN', 'Departure Percentage: JUL', 'Departure Percentage: AUG', 'Departure Percentage: SEP', 'Departure Percentage: JUN-SEPT']

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[departure_columns])
plt.title('Box Plot of Departure Percentages for Months')
plt.ylabel('Departure Percentage')
plt.xticks(rotation = 45)
plt.xlabel("Months")
plt.show()
```

## Violin Plot of Actual Rainfall for Months.

```python
rainfall_columns = ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT', 'Actual Rainfall: JUN-SEPT']

plt.figure(figsize=(12, 6))
sns.violinplot(data=df[rainfall_columns])
plt.title('Violin Plot of Actual Rainfall for Months')
plt.ylabel('Actual Rainfall')
plt.xticks(rotation = 45)
plt.xlabel("Months")
plt.show()
```

## Violin Plot for Departure Percentages for Months.

```python
departure_columns = ['Departure Percentage: JUN', 'Departure Percentage: JUL', 'Departure Percentage: AUG', 'Departure Percentage: SEP', 'Departure Percentage: JUN-SEPT']

plt.figure(figsize=(12, 6))
sns.violinplot(data=df[departure_columns])
plt.title('Violin Plot of Departure Percentages for Months')
plt.ylabel('Departure Percentage')
plt.xticks(rotation = 45)
plt.xlabel("Months")
plt.show()
```
