## Checking for Possible non-printable Characters using repr function
The `repr()` function in Python is a built-in function that returns a string representation of an object. It is often used for debugging and development purposes because it produces a string that, if passed to the `eval()` function, would recreate the original object with the same value.

When applied to a string, `repr()` returns a string containing a printable representation of the string itself, including any escape characters. This is especially useful when you want to inspect strings for non-printable characters or characters that might not be immediately visible.

Here's a simple example:

```python
string_with_unprintable = 'Hello\nWorld'
print("String with Normal Printing:",string_with_unprintable)
print("String with Repr Function Printing:",repr(string_with_unprintable))

Output:
String with Normal Printing: Hello
World
String with Repr Function Printing: 'Hello\nWorld'

```


In this example, the `\n` in the string is a newline character. The `repr()` function will represent it with the actual escape sequence, making it clear that there is a newline in the string.

In the context of the previous issue, using `repr()` on the column names was helpful to reveal any unprintable characters or special characters that might not be immediately visible when printing the column names directly. This can be particularly useful when dealing with data cleaning and debugging tasks.

````python
# Print the column names with repr to show non-printable characters
print([repr(col) for col in df.columns])
````

## Removing Unprintable Characters

````python
df.columns = df.columns.str.replace('\xa0', '')

print("Columns after removing Unpredictable Characters:")
print([repr(col) for col in df.columns])
````

## We need to remove $ symbol from columns Adjusted gross (in 2022 dollars) and Actual gross.

````python
# Replace symbols in a specific column
remove_symbol = lambda col: df[col].replace('[\$,]', '', regex=True).replace('[\[].*?[\]]', '', regex=True)

# Apply the lambda function to all the columns
for i in df.columns:
  df[i] = remove_symbol(i)

# Printing the Dataset
print("After Removing Symbols from the Dataframe:\n")
print(df)
````

## Now Compared to the dataset the missing values are many, so instead of removing them we will replace them with the mean but since this is a string column we will first replace NaN with 0 then convert to numeric and then finally we can replace with mean.

````python
# Since Peak is no a numeric column we cannot use mean with that so will directly replace with 0 instead of NaN values
df['Peak'].fillna(0, inplace = True)

# Since All time Peak is no a numeric column we cannot use mean with that so will directly replace with 0 instead of NaN values
df['All Time Peak'].fillna(0, inplace = True)
````

## Printing Columns after Removing Symbol and Convering to Integer Datatype.

````python
# Converting to Numeric Data Type
df['Adjustedgross (in 2022 dollars)'] = df['Adjustedgross (in 2022 dollars)'].apply(pd.to_numeric)
print("\nColumn Adjusted gross (in 2022 dollars) is:\n")
print(df['Adjustedgross (in 2022 dollars)'])

# Converting to Numeric Data Type
df['Actualgross'] = df['Actualgross'].apply(pd.to_numeric)
print("\nColumn Actual gross is:\n")
print(df['Actualgross'])

# Converting to Numeric Data Type
df['Peak'] = df['Peak'].apply(pd.to_numeric)
print("\nColumn Peak is:\n")
print(df['Peak'])

# Converting to Numeric Data Type
df['All Time Peak'] = df['All Time Peak'].apply(pd.to_numeric)
print("\nColumn All Time Peak is:\n")
print(df['All Time Peak'])
````

## Replacing Missing Values in Peak and All Time Peak with Mean

````python
# Since Peak is no a numeric column we cannot use mean with that so will directly replace with 0 instead of NaN values
df['Peak'].replace(0,df['Peak'].mean(), inplace = True)

# Since All time Peak is no a numeric column we cannot use mean with that so will directly replace with 0 instead of NaN values
df['All Time Peak'].replace(0, df['All Time Peak'].mean(), inplace = True)

print("Column Peak is:\n")
print(df['Peak'])

print("\nColumn All Time Peak is:\n")
print(df['All Time Peak'])
````

## We can see that Their is Huge Difference between Scales of other Columns and Gross so we will Normalize Actual and Adjusted Gross

````python
# Normalizing Actual and Adjusted Gross before Visualizing

# Normalising Actual Gross

min_val = df['Actualgross'].min()
max_val = df['Actualgross'].max()

# Define your new min and max
new_min = 10000
new_max = 50000

# Apply normalization
df['Actualgross'] = ((df['Actualgross'] - min_val) / (max_val - min_val)) * (new_max - new_min) + new_min

print("Actual Gross after Normalization is:\n")
print(df['Actualgross'])

# Normalising Adjusted Gross

min_val = df['Adjustedgross (in 2022 dollars)'].min()
max_val = df['Adjustedgross (in 2022 dollars)'].max()

# Define your new min and max
new_min = 10000
new_max = 50000

# Apply normalization
df['Adjustedgross (in 2022 dollars)'] = ((df['Adjustedgross (in 2022 dollars)'] - min_val) / (max_val - min_val)) * (new_max - new_min) + new_min

print("\nAdjusted Gross after Normalization is:\n")
print(df['Adjustedgross (in 2022 dollars)'])
````

## First we will remove symbol and numbers from column names

````python
# Clean the column names by removing leading and trailing whitespaces
df.columns = df.columns.str.strip()
````

## Checking for Possible non-printable Characters using repr function

````python
# Print the column names with repr to show non-printable characters
print([repr(col) for col in df.columns])
````

## Since we have not found any non-printable character we can now remove unneccesary symbols from data.

````python
df.columns = df.columns.str.replace('>>', '')

print("Columns after removing Unwanted Characters:")
print([repr(col) for col in df.columns])
````

## Now we will rename Columns which are labeled as Unnamed as Column 1 2 etc.

````python
df.rename(columns = {'Unnamed: 2' : 'Column 1', 'Unnamed: 3' : 'Column 2', 'Unnamed: 4' : 'Column 3', 'Unnamed: 7' : 'Column 4', 'Unnamed: 8' : 'Column 5', 'Unnamed: 9' : 'Column 6', 'Unnamed: 12' : 'Column 7', 'Unnamed: 13' : 'Column 8', 'Unnamed: 14' : 'Column 9'}, inplace = True)

# Printing Data after Renaming Columns
# Printing First 7 rows of dataset
print("Dataset is:\n")
print(df.head(n = 7))

# Printing Columns after Renaming
print("\nColumns after Renaming:\n")
print(df.columns)
````

## Since many columns are of object data type we will first fill na values with 0 and then convert them to numeric and then imputation can be done.

````python
# Replacing na values in dataset
df.fillna(0, inplace = True)
````

## For Columns seeing if any categorical value is given

````python
# Apply the lambda function to all the columns
for i in df.columns:
  print(f"\nUnique Values in Dataframe Column {i} is:\n")
  print(df[i].value_counts())
````

## We Can see that no Categorical Value is present just some strings are there so we can directly remove strings and then convert to numeric values.

````python
# Function to convert columns to numeric
Convert_Numeric = lambda col: df[col].apply(pd.to_numeric, errors = 'coerce')

# Apply the lambda function to all the columns
for i in df.columns:
  if i != 'Segment':
    df[i] = Convert_Numeric(i)

# Since some strings were present it will be converted to NaN and we need to fill it
df.fillna(0, inplace = True)

# Printing the Dataset
print("After Conversion to Numeric Values Dataframe is:\n")
print(df)
````

## Since there are many missing value column in dataset we cannot directly remove them, so we will impute with mean.

````python
# Apply the lambda function to all the columns
for i in df.columns:
  if i != 'Segment':
    df[i] = df[i].replace(0, df[i].mean())

# Printing the Dataset
print("After Conversion to Numeric Values Dataframe is:\n")
print(df)
````
