Importing pyplot from matplotlib

```python
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib as mat
```

Plotting directly from plot function<br>
Passing one dimensional data

```python
plt.plot([1,5,3,6,9,7])
# For x labeling
plt.xlabel("X axis")
# For y labeling
plt.ylabel("Y axis")
# For title of the plot
plt.title("One Dimensional data graph")
# Legend in plot
plt.legend(["Data"])
# To print the plot or to show the plot
# Without this argument it would not be displayed
plt.show()
```

Passing two dimensional data

```python
plt.plot([[1,5,3,6,9,7],[12,5,4,3,6,1],[1,5,3,6,9,7]])
# Labelling in graph
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("Index and Age relationship")
plt.legend(["Age Line","Random Value","X1","X2","X3","X4"])
plt.show()
```

Doesn't work with 3 d data<br>
plt.plot([[[1,5,3,6,9,7],[12,5,4,3,6,1],[1,5,3,6,9,7]]])

Plotting in form of point plot red color line  point hexagon shape dashed<br>
linewidth and marker size is used to specify size to graph

```python
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.",linewidth=5,markersize=20)
# Labeling
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("Index and Age relationship")
plt.legend(["Age Line","Random Value"])
plt.show()
```

To get details about markers and line points to be used

```python
print("Information about plot function is:\n")
help(plt.plot)
```

Directly plot scatter plot<br>
we set size of points in s

```python
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=[10,20,30,40,50,60]) # Give values in s or directly pass through range
plt.show()
```

Giving markers by using mat.markers.CARETDOWNBASE<br>
Setting the size using range

```python
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(10,601,100),c=range(1,7),marker=mat.markers.CARETDOWNBASE)
plt.xlabel("Amount(in lacs)")
plt.ylabel("Age")
plt.title("Amount and Age relationship")
plt.legend(["Age Line"])
plt.show()
```

 To get details about markers and line points to be used

```python
print("Information about scatter function is:\n")
help(plt.scatter)
```

subplot function takes three arguments rows,columns,index<br>
First we write subplot function then gives graph

```python
plt.subplot(1,2,1)
# Plot 1
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(10,601,100),c=range(1,7)) #marker=mat.markers.CARETDOWNBASE)
plt.xlabel("Amount(in lacs)")
plt.ylabel("Age")
plt.title("Amount and Age relationship Scatter Plot")
# Plot 2
plt.subplot(1,2,2)
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.",linewidth=5,markersize=20)
plt.xlabel("Amount(in lacs)")
plt.ylabel("Age")
plt.title("Amount and Age relationship Line plot")
# Displaying the plot
# Title to entire fig
plt.suptitle("Amount and Age relationship")
plt.show()
```

Creating dataframe

```python
df1=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],index=['a','b','c'],columns=['d','e','f'])
print("Dataframe after index and columns labels made by using list is:")
print(df1)
```

Plotting

```python
# plt.subplot(2,2,1) #-By default it takes the first graph in first index
df1.hist("d","e","f")
plt.subplot(2,2,2)
df1.boxplot()
plt.show()
```

Creating 4 plots together

```python
plt.subplot(2,2,1)
# Plot 1
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(10,601,100),c=range(1,7))
plt.xlabel("Amount(in lacs)")
plt.ylabel("Age")
plt.title("Amount and Age relationship Scatter Plot")
# Plot 2
plt.subplot(2,2,2)
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.",linewidth=5,markersize=20)
plt.xlabel("Amount(in lacs)")
plt.ylabel("Age")
plt.title("Amount and Age relationship Line plot")
# Plot 3
plt.subplot(2,2,3)
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(10,601,100),c=range(1,7),marker=mat.markers.CARETDOWNBASE)
plt.xlabel("Amount(in lacs)")
plt.ylabel("Age")
plt.title("Amount and Age relationship")
plt.legend(["Age Line"])
# Plot 4
plt.subplot(2,2,4)
df1.boxplot()
plt.show()
```
