# Create a menu driven program to insert, display, and delete elements using arrays. Insertion and deletion has three possible cases:(i) insertion at beginning, (ii) insertion at the end, and  (iii) insertion in the middle. Also, include linear search and binary search functions in the menu.



```python
# Creating function for displaying array
def display(arr):
    print("Array is:")
    for i in range(0,len(arr)):
        print("Element",i+1,"is:",arr[i])

# Creating Function for Insertion

# Beginning
def insert_begin(arr,ele):
        arr.insert(0,ele)
        print("Element inserted at beginning")
        display(arr)
    
# End
def insert_end(arr,ele):
    arr.append(ele)
    print("Element inserted at beginning")
    display(arr)

# At any place(Index)
def insert_anyplace(arr,ele,ind):
    arr.insert(ind-1,ele)
    print("Element inserted at Index:",ind)
    display(arr)
    
# Creating Function for Deletion

# Beginning
def delete_begin(arr):
    arr.pop(0)
    print("Element deleted from beginning")
    display(arr)
    
# End
def delete_end(arr):
    arr.pop()
    print("Element deleted from end")
    display(arr)

# At any place(Index)
def delete_anyplace(arr,ind):
    arr.pop(ind-1)
    print("Element deleted from Index",ind)
    display(arr)
    
# Searching

# Linear search
def linearSearch(arr,k):
    for i in arr:
        if i==k:
            print(k,"is present at:",arr.index(i)+1)
            break
    else:
        print(k,"is not present in array.")
        
# Binary search
def binarySearch(arr,k,s,e):
    if e >= s:
        m = (s+e) // 2
        if arr[m]== k:
            return m
        elif arr[m]>k:
            return binarySearch(arr,k,s,m-1)
        else:
            return binarySearch(arr,k,m+1,e)   
    else:
        return -1
        
# Menu function

def menu(arr):
    op=int(input("Enter 1 for insertion, 2 for Deletion, 3 for display,4 for linear search(Unsorted array),5 for binarysearch(Sorted array): "))
    
    if op==1:
        el=int(input("Enter element to insert: "))
        ch=int(input("Enter 1 for insertion in beginning, 2 for insertion in end and 3 for insertion at any place: "))
        
        if ch==1:
            insert_begin(arr,el)
        elif ch==2:
            insert_end(arr,el)
        else:
            ind=int(input("Enter index to insert: "))
            insert_anyplace(arr,el,ind)
            
    elif op==2:
        ch=int(input("Enter 1 for deletion from beginning, 2 for deletion from end and 3 for deletion from any place: "))
        
        if ch==1:
            delete_begin(arr)
        elif ch==2:
            delete_end(arr)
        else:
            ind=int(input("Enter index to delete: "))
            delete_anyplace(arr,ind)
            
    elif op==3:
        display(arr)
        
    elif op==4:
        el=int(input("Enter element to search: "))
        linearSearch(arr,el)
        
    elif op==5:
        l=len(arr)-1
        arr.sort()
        print("Sorted array is: ",arr)
        el=int(input("Enter element to search: "))
        b=binarySearch(arr,el,0,l) 
        if b != -1:
            print(el,"is present at:",int(b)+1)
        else:
            print(el,"is not Present")   
    
    else:
        print("Invalid Case!!")
        
# If we want to repeat functions,for that choice_repeat function
def choice_repeat(arr):
    
    while int(input("Enter 1 for repeating functions on Array or any other key to exit: "))==1:
        menu(arr)
    else:
        print("Program Executed Successfully")
        
# Main function

if __name__=='__main__':
    
    arr=[]
    n=int(input("Enter Number of elements: "))
    
    for i in range(0,n):
        ele=int(input("Enter element: "))
        arr.append(ele)
   
    menu(arr)
    
    # Calling choice_repeat function
    choice_repeat(arr)
```

# Implement Queue data structure and associated operations using arrays. Design a menu for using the operations.



```python
# Creating function for displaying Queue
def display(arr):
    print("Queue is:")
    for i in range(0,len(arr)):
        print("Element",i+1,"is:",arr[i])

# Enqueue-Adding element in Queue(in the end)
def Enqueue(arr,ele):
    arr.append(ele)
    print("Element added in Queue")
    display(arr)
    
# Dequeue-Deleting element in Queue(in the front)
def Dequeue(arr):
    arr.pop(0)
    print("Element removed from Queue")
    display(arr)

# Front-Element at front/start of Queue
def front(arr):
    print("Front element of Queue is:",arr[0])
    
# Rear-Element at end/back of Queue
def rear(arr):
    print("Rear element of Queue is:",arr[len(arr)-1])
    
# Checking if Queue is empty or not
def isEmpty(arr):
    if not arr:
        print("Queue is Empty")
    else:
        print("Queue is not Empty")
        
# Checking if Queue is full or not
# We need to set a maximum size for Queue,if the size of Queue is equal to that size we say Queue is full
# Here the max size is n+12 where n is number of elements in Queue
def isFull(arr):
    if len(arr)==n+12:
        print("Queue is Full")
    else:
        print("Queue is not Full")
        
# Menu function

def menu(arr):
    op=int(input("Enter 1 for Enqueue, 2 for Dequeue, 3 for display,4 for front,5 for rear,6 for isEmpty or 7 for isFull: "))
    
    if op==1:
        el=int(input("Enter Element to insert: "))
        Enqueue(arr,el)
            
    elif op==2:
        Dequeue(arr)
            
    elif op==3:
        display(arr)
        
    elif op==4:
        front(arr)
        
    elif op==5:
        rear(arr)
        
    elif op==6:
        isEmpty(arr)
    
    elif op==7:
        isFull(arr)
    
    else:
        print("Invalid Case!!")
        
# If we want to repeat functions,for that choice_repeat function
def choice_repeat(arr):
    
    while int(input("Enter 1 for repeating functions on Queue or any other key to exit: "))==1:
        menu(arr)
    else:
        print("Program Executed Successfully")
        
# Main function

if __name__=='__main__':
    
    arr=[]
    n=int(input("Enter Number of elements: "))
    
    for i in range(0,n):
        ele=int(input("Enter element: "))
        Enqueue(arr,ele)
   
    menu(arr)
    
    # Calling choice_repeat function
    choice_repeat(arr)
```
