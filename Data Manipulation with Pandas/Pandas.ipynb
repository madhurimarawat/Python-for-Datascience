```python
names=pd.Series(['mark','Rita','vicki','io'],
                 index=[1,2,3,'d'])
# Accessing elements of lists
# Done with Indexing-Starts from 0 but we have explicitly started it from 1 by setting in index parameter
print("Series created from list is:")
print(names)
print('\n')
first_element=names[1]
print('First element:',first_element)
print('\n')
# Accessing by giving index labels
print("Element at index label 'd' is:",names['d'])
print('\n')
# Slicing: start:stop:step
print("From index 1 to index 4 i.e element 2 to 4 is:")
print(names[1:4])
```

```python
names=pd.Series(['mark','Rita','vicki','io','michelis','jane','doe'],
                 index=[1,2,3,'d','e','f','g'])
print("Series created from list is:")
print(names)
print('\n')
# Slicing: start:stop:step
print("From index 1 to index 4 i.e element 2 to 4 is:")
print(names[1:3])
print('\n')
# Accessing by giving index labels
print("From index label 'd' to index label 'g' is:")
print(names['d':'g'])
```

```python
names=pd.Series(['mark','Rita','vicki','io','michelis','jane','doe'],
                 index=['a','b','c','d','e','f','g'])
print("Series created from list is:")
print(names)
print('\n')
# Accessing by giving index labels
print("From index label 'b' to index label 'e' is:")
print(names['b':'e'])
print('\n')
# Accessing by giving index labels and index
# Accessing multiple elements
print("Elements at index 2,3 and 0 are:")
print(names[[2,3,0]])
print('\n')
print("Elements at index label a,g and f are:")
print(names[['a','g','f']])
# Updating 3 element
names[2]='Sanjay'
print("After Updating Series is:")
print(names)
```

```python
# Creating several lists and then series and combining them in dataframe
# Creating List
author=['Jitendar','purnima','mahadevi','premchand','nisha','Hello']
print("Author List is:\n",author)
print('\n')

# Creating List
article=['210',220,234,567,89,90]
print("Article List is :\n",article)
print('\n')

# Creating Series
au_series=pd.Series(author)
print("Author-series is:")
print(au_series)
print('\n')

# Creating Series
ar_series=pd.Series(article)
print("Article-series is:")
print(ar_series)
print('\n')

# Combining in frame variable
frame={'Author':au_series,'Article':ar_series}
# Creating Dataframe
result=pd.DataFrame(frame)
print("Dataframe made by combining lists and then series is:")
print(pd.DataFrame(frame))
```

```python
#  Rows and Columns can be accessed by giving name in [[]]
# copy() function is used to copy
new=result[['Author']].copy()
print("Copied column is:")
print(new)
print("Type of Copied column is:",type(new))
```

```python
# Creating dataframe
student_de={'Name':['Tom','James','Ricky','Vin','Steve','Lee','Betina'],'Name1':['Tom','James','Ricky','Vin','Steve','Lee','Betina'],
   'Age':[23,56,34,23,52,65,43],'Rating':[2.89,6.34,3.9,6.78,4.9,8.9,9.0]}
indexl=['r1','r2','r3','r4','r5','r6','r7']
df=pd.DataFrame(student_de,index=indexl)
print("Dataframe is:")
print(df)
# Adding new column in dataframe
# Since it is string it directly concatenate it
df['Sum']=df['Name']+df['Name1']
print("Dataframe adding new column is:")
print(df)
```

```python
print("Dataframe is:")
print(df)
print("\n")
# Finding sum of two columns
col_list=['Age','Rating']
# Adding new column
df['Sum1']=df[col_list].sum(axis=1)
print("Dataframe adding new column is:")
print(df)
```

```python
print("Dataframe is:")
print(df)
print("\n")
# Using iloc function
# now since axis=1 it search in columns and from 1 index column is string so it takes 2 index column and print it as it is
df['Sum2']=df.iloc[:,1:3].sum(axis=1)
print("Updated Dataframe is:")
print(df)
```

```python
print("Dataframe is:")
print(df)
print(("\n"))
# Finding sum
# In iloc first : is for all rows and index 1 is 2 column and index 4 is 5 column and this is why it added(Concatenated) them
df['Sum3']=df.iloc[: ,[1,4]].sum(axis=1)
print("Updated Dataframe is:")
print(df)
```

```python
# Descriptive statistics
# Finding mean
# Age column
print("Age column is:")
print(df['Age'])
mean=df['Age'].mean()
print("Mean is:",mean)
print("Median is:",df['Age'].median())
print("Mode is:",df['Age'].mode())
print("Count is:",df['Age'].count())
print("Standard Deviation is:",df['Age'].std())
```

```python
# To delete column we use drop function
# Give column name to be dropped in columns
# Dosen't update in actual dataframe till inplace is set to be True
print(" Original Dataframe is:")
print(df)
print("\n")
df.drop(columns=['Sum2'],inplace=True)
print(" Updated Dataframe is:")
print(df)
```

```python
df=pd.DataFrame({'name':['Emily','Liam','Lisa','Adam','lina','dany'],'current':[98,12,85,14,78,90],'overall':[184,258,789,456,123,457],'group':["A","A","B","A","D","C"]})
print("Dataframe is:")
print(df)
```

```python
# Using ranking methods
df["rank_default"]=df["group"].rank()
print(df)
```

```python
# Using ranking methods
df["rank_default_desc"]=df["group"].rank(ascending=False)
df=df.sort_values(by="rank_default_desc",ignore_index=True)
print(df)
```

```python
# create rank columns
df["rank_default"]=df["group"].rank(ascending=False)
df["rank_min"]=df["group"].rank(method="min",ascending=False)
df["rank_max"]=df["group"].rank(method="max",ascending=False)
print(df)
```

```python
# Sort rows
df=df.sort_values(by="rank_default",ignore_index=True)
print(df)
```

```python
# Sort rows
df=df.sort_values(by="rank_default",ignore_index=False)
print(df)
```

```python
# filter() function is used to filter dataframe based on condition
# axis 0 is for rows and axis 1 is for columns
df2=df.filter(items=['name','current'])
print("Filtering based on name and current is:")
print(df2)
```

```python
print("Dataframe is:")
print(df,"\n")
# where function gives rows which satisfy condition and replaces rest of the values with NaN(Not a Number)
# Storing in a new dataframe
df2=df.where(df.overall>200)
print("Student with overall greater than 200 are:")
print(df2)
```

```python
print("Dataframe is:")
print(df,"\n")
# we can also give argument to replace values with instead of NaN
df2=df.where(df.overall>200,'GOOD')
print("Student with overall greater than 200 in which rest of values replaced with GOOD are:")
print(df2)
```

```python
print("Dataframe is:")
print(df,"\n")
# Applying two conditions using where function and bitwise &
cond1=df.overall>200
cond2=df.overall<300
df2=df.where(cond1 &cond2,other="TRIAL")
print("Student with overall greater than 200 and lesser than 300 in which rest of values replaced with TRIAL are:")
print(df2)
```

```python
print("Dataframe is:")
print(df,"\n")
# query function is used to find information about a particular name in row or column
print("Information about Liam is:")
print(df.query("name=='Liam'"),"\n")
print("Information about current 90 is:")
print(df.query("current==90"),"\n")
# Taking variable
value=90
print("Information about current 90 using variable is:")
print(df.query("current==@value"),"\n")
print("Information about current 98 and 14 is:")
df.query("current in [98,14]")
```

```python
print("Dataframe is:")
print(df,"\n")
# Since no data satisfy the condition it prints empty dataframe
# Using query function for multiple conditions
print("Student with current greater than 90 and overall greater than 200 are:")
df.query("current>90 & overall>200")
```
